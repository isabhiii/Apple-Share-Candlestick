```python
mpf.plot(data['2020-12' : '2021-11'], type= 'candle',mav= (20), volume= True)
```

```python
mpf.plot(data['2020-12' : '2021-11'], type= 'candle',mav= (20), volume= True, tight_layout= True)
```

```python
# Assigning the Title
mpf.plot(data['2020-12' : '2021-11'], type= 'candle', title= 'Apple Share Price 2020/21',mav= (20), volume= True, tight_layout= True)
```

```python
mpf.plot(data['2020-12' : '2021-11'], figratio= (20,15),type= 'candle', title= 'Apple Share Price 2020/21',mav= (20), volume= True, tight_layout= True)
```

```python
# Changing the style to Yahoo
mpf.plot(data['2020-12' : '2021-11'], figratio= (20,15),type= 'candle', title= 'Apple Share Price 2020/21',mav= (20), volume= True, tight_layout= True, style= 'yahoo')
```
